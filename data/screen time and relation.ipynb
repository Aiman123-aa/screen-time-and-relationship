```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()

x = np.array([1.5,1,3,2])
y = np.array([3.50,2,1,1])



ax.scatter(x, y, color='blue')


ax.set_xlabel("Parent's screen time (hours)")
ax.set_ylabel("Children's screen time (hours)")
ax.set_title("Screen use")



plt.show()

```

```python
ax.scatter(x, y, color='blue', label="Screen Time Relationship")


ax.set_xlabel("Parent's Screen Time (hours)")
ax.set_ylabel("relationship_conflict")
ax.set_title("Parent's screen time vs relationship conflict")
ax.legend()


plt.show()
```

```python
ax.scatter(x, y, color='blue', label="Screen Time & Relationship")


ax.set_xlabel("relationship_conflict")
ax.set_ylabel("Parent's Screen Time (hours)")
ax.set_title("Parent's screen time vs relationship conflict")
ax.legend()


plt.show()
```
